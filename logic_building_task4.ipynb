```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.\n")
            continue

        if amount not in valid_plans:
            print("Invalid plan selected. Available plans:", valid_plans)
            print("Please try again.\n")
            continue

        print("Recharge successful for ₹", amount)
        break

# Call the function
validate_recharge()

#output
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")

# Sample Data
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 8,
    "Monitor": 25
}

check_inventory(inventory)

#output
```

```python
# Problem 3: Student Result Processing System

def student_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Sample Data
marks_list = [60, 45, 70, 55]

result = student_result(marks_list)
print("Result:", result)

#output
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak == "yes":
        fare += fare * 0.25

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break

#output
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Sample Data
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

status = check_attendance(attendance)
print("Attendance Status:", status)


#output
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character @#$)"

# Sample Test
pwd = input("Enter password: ")
print(check_password(pwd))

#output
```
