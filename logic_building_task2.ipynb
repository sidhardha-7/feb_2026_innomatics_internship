```python
#Problem 2: Highest Salary from Employee Data

print("\nProblem 2 - Highest Salary")

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_salary = 0
highest_employee = ""

for name in employees:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)

#output
```

```python
# Problem 1: Unique Words in a Sentence


print("Problem 1 - Unique Words")

sentence = input("Enter a sentence: ")

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)

#output
```

```python
# Problem 3: Find Maximum and Minimum Values

print("\nProblem 3 - Max and Min")

numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)

#output
```

```python
# Problem 4: Count Products Above Price Threshold

print("\nProblem 4 - Products Above 1000")

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)

#output
```

```python
# Problem 5: Calculate Attendance Percentage

print("\nProblem 5 - Attendance Percentage")

attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)

#output
```

```python
# Problem 6: Remove Duplicate Phone Numbers

print("\nProblem 6 - Remove Duplicate Phone Numbers")

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = list(set(phone_numbers))

print("Unique phone numbers:", unique_numbers)

#output
```

```python
# Problem 7: Count Character Frequency

print("\nProblem 7 - Character Frequency")

text = input("Enter a string: ")

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Character Frequency:", frequency)

#output
```

```python
# Problem 10: Calculate Average Marks

print("\nProblem 10 - Average Marks")

marks = [70, 75, 80, 65]

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print("Average Marks:", average)

#output
```
