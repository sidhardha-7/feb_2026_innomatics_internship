```python
#Problem 1: Social Media – Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0
    
    for likes in likes_list:
        total_likes += likes
    
    print("Total Likes:", total_likes)
    
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example
analyze_engagement([200, 300, 150, 400])

#output
```

```python
#Problem 2: Healthcare – Medicine Stock Alert System

def check_medicine_stock(stock):
    print("Medicine Stock:", stock)
    
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


# Example
check_medicine_stock(6)

#output
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker

def check_rainfall(rainfall_data, required_level):
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", average)
    
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example
check_rainfall([70, 75, 68, 80, 67], 70)

#output
```

```python
#Problem 4: Social Media – Duplicate Account Detection

def check_duplicate_accounts(usernames):
    unique_usernames = set(usernames)
    
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example
check_duplicate_accounts(["luna123", "ravi01", "luna123"])

#output
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices):
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)


# Example
filter_premium_crops([1500, 2500, 1800, 3200])

#ouput
```

```python
#Problem 7: System Monitoring – Application Health Checker

def check_application_health(errors):
    print("Error Count:", errors)
    
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example
check_application_health(7)

#output
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker

def check_transaction(amount):
    daily_limit = 50000
    
    print("Transaction Amount:", amount)
    
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example
check_transaction(60000)

#output
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System

def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    
    for status in attendance_list:
        if status == "Present":
            attended += 1
    
    percentage = (attended / total_classes) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example
check_attendance(["Present", "Absent", "Present", "Present", "Present"])

#output
```

```python
#Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status


# Example
total_bill, usage_status = calculate_electricity_bill(250)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)

#output
```
